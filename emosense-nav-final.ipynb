<h2 style="text-align:center;background-color:#19376D; color:White; font-size: 20px">Human emotions come in wide varieties and intensities. It becomes quite tough to understand  them just by relating the context to a emotion, thus failing to get the underlying thought of the narrator. This problem becomes more prominent when a written text is only way to convey the thoughts of the narrator or it is only way to express ones feeling. A way must be devised to understand complex emotions.
</h2>

<h1 style="text-align:center;background-color:#0E153A; color:White; font-size: 20px">

We are using the three independent variables (Valence,Arousal,Dominance) for our model prediction
    
'Valence' measures how positive or negative the emotion is. A positive valence score indicates that the emotion is positive, while a negative valence score indicates that the emotion is negative. A score of zero indicates a neutral emotion.

'Arousal' measures the intensity of the emotion. A Positive arousal score indicates that the emotion is intense, while a negative arousal score indicates that the emotion is less intense. A score of zero indicates a neutral level of arousal.

'Dominance' measures how in control a person feels in a given situation. A positive dominance score indicates that the person feels in control, while a negative dominance score indicates that the person feels less in control. A score of zero indicates a neutral feeling of control.
    
</h1>

<h2 style="text-align:center;background-color:#0E157A; color:White; font-size: 40px">A Three-Factor Theory of Emotions </h2>

#### The paper titled "Evidence for a Three-Factor Theory of Emotions" presents empirical evidence supporting the validity of the Three-Factor Theory. The study includes a series of experiments that examine the relationship between physiological arousal, cognitive appraisal, and emotional labeling. The findings demonstrate that individuals experiencing similar physiological arousal interpret and label their emotions differently based on cognitive appraisal and situational cues. This research provides strong support for the Three-Factor Theory and contributes to our understanding of the complex nature of emotions. You can access the paper here: [Evidence for a Three-Factor Theory of Emotions](https://www.researchgate.net/publication/222741832_Evidence_for_a_Three-Factor_Theory_of_Emotions)

### How the Emotions are distributed in EmoBank Dataset?

```python
emos = pd.read_csv('/kaggle/working/EmoSense/EmoBank_emotions.csv')
emos.head()
```

```python
print("Number of unique emotions:", emos.emotions.nunique())
```

```python
# Calculate the frequency of each emotion
emotion_counts = emos['emotions'].value_counts()

# Create a word cloud based on the frequency of emotions
wordcloud = WordCloud(width=800, height=400, background_color='black').generate_from_frequencies(emotion_counts)

# Create a larger plot
plt.figure(figsize=(14, 10))

# Add the word cloud image with a dark background
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Emotion Frequency Word Cloud", color='white', fontsize=16)
plt.savefig('wordcloud.png', dpi=300)
plt.show()
```

```python
top_10_emotions = emotion_counts.head(10)
plt.figure(figsize=(10, 6))
bars = plt.bar(top_10_emotions.index, top_10_emotions.values, zorder=2)
mlp.add_bar_gradient(bars=bars)
plt.xlabel('Emotions')
plt.ylabel('Counts')
plt.title('Top 10 Emotions and Their Counts')
plt.xticks(rotation=45)
plt.show()
```
