# Factor Analysis (NMF) :

![Factor Analysis](https://jalammar.github.io/images/explaining/neuron-factors-decomposition.PNG)

#### Advantages:
1. The factors can be achiieved by decomposing the matrix holding the activations values of FFNN neurons using Non-negative Matrix Factorization(NMF)[1].
2. Beyond dimensionality reduction, Non-negative Matrix Factorization can reveal underlying common behaviour of groups of neurons. It can be used to analyze the entire network, a single layer, or groups of layers[1].

## Pre-trained DIstilBERT Model Neuron Factorization:

```python
lm = ecco.from_pretrained('distilbert-base-uncased', activations=True)
```

```python
text = ''' The vast expanse of the ocean stretches out before me, its calm waters reflecting the sky above.
Its tranquil surface, undisturbed by the chaos of the world, inspires a sense of serene tranquility.
The soft whisper of the waves kissing the shoreline is a peaceful symphony, a melodious song of
nature that soothes the soul. The gentle, rhythmic motion of the ocean’s currents embodies a silent
strength, a sense of enduring serenity that is profoundly peaceful. The boundless beauty of the sea,
with its captivating calmness, offers a sanctuary from the bustling noise of everyday life, a haven
where one can find peace and solitude. It’s a place where the mind can rest, allowing thoughts to
flow as freely as the waters themselves. The ocean, in its infinite tranquility, serves as a reminder
of the simple and quiet beauty that exists in our world.
'''
# Factorize the activations of all layers
inputs = lm.tokenizer([text], return_tensors="pt")
output = lm(inputs)
nmf_1 = output.run_nmf(n_components=8)                         #non-negative matrix factorization on 8 components
nmf_1.explore()
```

> - Above visualization shows how a pretrained DistilBERT model's prediction activated by input sentence tokens using dimention reduction.
> - A group of neurons in DistilBERT tend to fire in response to commas and other punctuation. Other groups of neurons tend to fire in response to pronouns. we use this visualization to factorize neuron activity in individual FFNN layers.

## Our Finetuned Model V Activations:

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
model_V = AutoModelForSequenceClassification.from_pretrained("/kaggle/input/model-new-up/MODELs/MODEL_V",output_attentions=True)
tokenizer_V = AutoTokenizer.from_pretrained("/kaggle/input/model-new-up/MODELs/MODEL_V")
model_V
```

```python
model_config_V = {
    'embedding': "embeddings.word_embeddings",
    'type': 'mlm',
    'activations':['ffn\.lin2'],
    'token_prefix': '',
    'partial_token_prefix': '##'
}
V_model_id = "/kaggle/input/model-new-up/MODELs/MODEL_V"
```

```python
ecco_model = ecco.from_pretrained(V_model_id, 
                                  activations=True,
                                  model_config=model_config_V)
```

### Factorizing Activations of a All Layer Layers

```python
# Factorize the activations of all layers
inputs = ecco_model.tokenizer([text], return_tensors="pt")
output = ecco_model(inputs)
nmf_1 = output.run_nmf(n_components=8)                           #non-negative matrix factorization on 8 components
nmf_1.explore()
```

### Factorizing Activations of a Single Layer

```python
# Factorize the activations of only the first layer
nmf_2 = output.run_nmf(n_components=8, from_layer=0, to_layer=1)
nmf_2.explore()
```

> - Factors that focus on specific portions of the text (beginning, middle, and end. This is interesting as the only signal the model gets about how these tokens are connected are the positional encodings. This indicates the neurons that are keeping track of the order of words. Further examination is required to assess whether these FFNN neurons directly respond to the time signal, or if they are responding to a specific trigger in the self-attention layer's transformation of the hidden state.
> - Factors corresponding to linguistic features, for example, we can see factors for pronouns, auxiliary verbs, other verbs and noun such as **"ocean"**,**"waters"**, linking verbs **(is, are,** were), and a factor that favors nouns and their adjectives.
> - A factor corresponding to commas, a syntactic feature.
> - We can observe that the model concentrates on such words that are related to emotions, though this is human intuition as the model never sees the actual emotions. words like **"tranquil", "peaceful", "solitude", "symphony"**

# Visualizing Attentions in Our Finetuned Model:
 - multiple views that each offer a unique lens into the attention mechanism.[2]

```python
from bertviz import head_view, model_view
from bertviz.neuron_view import show
from IPython.display import IFrame
```

## Model V(Finetuned DistilBERT):

```python
sentence_a = "It was a boring day"
sentence_b = "The ocean is so peaceful"
inputs = tokenizer_V.encode_plus(sentence_a, sentence_b, return_tensors='pt')
outputs = model_V(**inputs)
attention = outputs[-1]
tokens = tokenizer_V.convert_ids_to_tokens(inputs["input_ids"][0]) 
html_model_view = model_view(attention, tokens, html_action='return', display_mode="light")
html_head_view = head_view(attention, tokens, html_action='return')
# model_view(attention, tokens, display_mode="light")
with open('model_V_model_view.html', 'w') as file:
    file.write(html_model_view.data)
with open('model_V_head_view.html', 'w') as file:
    file.write(html_head_view.data)
html_file_path = ['model_V_model_view.html','model_V_head_view.html']
```
